```python
import numpy as np
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage.transform import resize

img_path = '/home/lingfeng/Desktop/Nested_CV/thesis_plots_drawing/coronal_images/6/6cor_1/6cor_1_114_1.jpg'

img_path_skull_stripping = '/home/lingfeng/Desktop/Nested_CV/thesis_plots_drawing/skull_stripping_images/6/6cor_1/6cor_1_114_1.jpg'

image = Image.open(img_path)

image_skull_stripping = Image.open(img_path_skull_stripping)

# print(image)

image = mpimg.imread(img_path)

image = resize(image, (256,256))

image_skull_stripping = mpimg.imread(img_path_skull_stripping)

f, axarr = plt.subplots(1,2) 

# print(image)

axarr[0].imshow(image[:,:,0])
axarr[1].imshow(image_skull_stripping[:,:,0])

plt.show()
```
