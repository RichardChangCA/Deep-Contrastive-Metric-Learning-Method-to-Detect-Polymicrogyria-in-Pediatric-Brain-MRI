```python
from skimage.util import montage
from glob import glob
import os
import nibabel as nib
import matplotlib.pyplot as plt

# Reference: https://www.kaggle.com/code/kmader/show-3d-nifti-images/notebook

images_folder = '/home/lingfeng/Desktop/Nested_CV/thesis_plots_drawing/skull_stripping'

all_images=glob(os.path.join(images_folder,'coronal.nii'))

test_image=nib.load(all_images[0]).get_data()

fig, ax1 = plt.subplots(1, 1, figsize = (20, 20))
ax1.imshow(montage(test_image), cmap ='bone')
fig.savefig('MRI_scan.png')
```
